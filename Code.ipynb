```python
import re
from collections import defaultdict, Counter
import time
import random
import string
import nltk
from nltk import bigrams, trigrams, FreqDist


def load_corpus_from_file(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        corpus = file.read().lower()
        corpus = re.sub(r'[^a-z]', '', corpus)  # Remove non-letters
    return corpus

corpus = load_corpus_from_file('corpus_training.txt')

def train_markov_chain(text, order=1):
    model = defaultdict(Counter)
    for i in range(len(text) - order):
        state = text[i:i+order]
        next_char = text[i+order]
        model[state][next_char] += 1
    all_chars = set(text)
    for state, next_chars in model.items():
        total = float(sum(next_chars.values())) + len(all_chars)
        for char in all_chars:
            next_chars[char] = (next_chars[char] + 1) / total  # Laplace smoothing
    return model

def train_markov_models(corpus, max_order=7):
    markov_models = {}
    for order in range(1, max_order + 1):
        markov_models[order] = train_markov_chain(corpus, order=order)
    return markov_models

markov_models = train_markov_models(corpus, max_order=7)
```

```python
def decrypt_with_hill_climbing(cipher_text, model, order=1, max_iterations=5000):
    def score_text(text):
        score = 0
        for i in range(len(text) - order):
            state = text[i:i+order]
            next_char = text[i+order]
            score += np.log(model[state].get(next_char, 1e-10))
        return score

    best_score = float('-inf')
    best_plain_text = cipher_text

    for _ in range(10):  # Multiple runs to avoid local optima
        best_key = list(string.ascii_lowercase)
        random.shuffle(best_key)
        best_key_map = dict(zip(string.ascii_lowercase, best_key))
        best_plain_text = ''.join(best_key_map.get(char, char) for char in cipher_text)
        best_score = score_text(best_plain_text)

        for iteration in range(max_iterations):
            i, j = random.sample(range(26), 2)
            new_key = best_key[:]
            new_key[i], new_key[j] = new_key[j], new_key[i]
            new_key_map = dict(zip(string.ascii_lowercase, new_key))
            new_plain_text = ''.join(new_key_map.get(char, char) for char in cipher_text)
            new_score = score_text(new_plain_text)

            if new_score > best_score:
                best_key, best_plain_text, best_score = new_key, new_plain_text, new_score

    return best_plain_text
```

```python
def decrypt_with_simulated_annealing(cipher_text, model, order=1, max_iterations=5000, initial_temp=100.0, cooling_rate=0.995):
    def score_text(text):
        score = 0
        for i in range(len(text) - order):
            state = text[i:i+order]
            next_char = text[i+order]
            score += np.log(model[state].get(next_char, 1e-10))
        return score

    def acceptance_probability(old_score, new_score, temperature):
        if new_score > old_score:
            return 1.0
        return np.exp((new_score - old_score) / temperature)

    best_key = list(string.ascii_lowercase)
    random.shuffle(best_key)
    best_key_map = dict(zip(string.ascii_lowercase, best_key))
    best_plain_text = ''.join(best_key_map.get(char, char) for char in cipher_text)
    best_score = score_text(best_plain_text)

    temperature = initial_temp

    for iteration in range(max_iterations):
        i, j = random.sample(range(26), 2)
        new_key = best_key[:]
        new_key[i], new_key[j] = new_key[j], new_key[i]
        new_key_map = dict(zip(string.ascii_lowercase, new_key))
        new_plain_text = ''.join(new_key_map.get(char, char) for char in cipher_text)
        new_score = score_text(new_plain_text)

        if acceptance_probability(best_score, new_score, temperature) > random.random():
            best_key, best_plain_text, best_score = new_key, new_plain_text, new_score

        # Cool down the temperature
        temperature *= cooling_rate

    return best_plain_text
```

```python
def mcmc_decrypt(cipher_text, model, order=1, max_iterations=5000):
    def score_text(text):
        score = 0
        for i in range(len(text) - order):
            state = text[i:i+order]
            next_char = text[i+order]
            score += np.log(model[state].get(next_char, 1e-10))
        return score
    
    def acceptance_probability(old_score, new_score):
        if new_score > old_score:
            return 1.0
        return np.exp(new_score - old_score)

    current_key = list(string.ascii_lowercase)
    random.shuffle(current_key)
    current_key_map = dict(zip(string.ascii_lowercase, current_key))
    current_plain_text = ''.join(current_key_map.get(char, char) for char in cipher_text)
    current_score = score_text(current_plain_text)
    
    best_plain_text = current_plain_text
    best_score = current_score

    for iteration in range(max_iterations):
        i, j = random.sample(range(26), 2)
        proposed_key = current_key[:]
        proposed_key[i], proposed_key[j] = proposed_key[j], proposed_key[i]
        proposed_key_map = dict(zip(string.ascii_lowercase, proposed_key))
        proposed_plain_text = ''.join(proposed_key_map.get(char, char) for char in cipher_text)
        proposed_score = score_text(proposed_plain_text)

        # Accept or reject the new key based on Metropolis-Hastings criterion
        if acceptance_probability(current_score, proposed_score) > random.random():
            current_key, current_plain_text, current_score = proposed_key, proposed_plain_text, proposed_score
        

        if current_score > best_score:
            best_plain_text, best_score = current_plain_text, current_score

    return best_plain_text
```

```python
import numpy as np 
import matplotlib.pyplot as plt

max_iterations = 1
def evaluate_accuracy(decrypted_text, original_text):
    correct = sum(1 for a, b in zip(decrypted_text, original_text) if a == b)
    return correct / len(original_text)

# Test and compare decryption methods for all Markov orders
def test_decryption_methods():
    plaintext = "thisisatestmessageforcryptanalysis"
    cipher_text, cipher_key = apply_substitution_cipher(plaintext)

    print(f"Ciphertext: {cipher_text}")
    print(f"Original key: {cipher_key}")

    # Test and compare decryption methods
    decryption_methods = {
        "Hill Climbing": decrypt_with_hill_climbing,
        "Simulated Annealing": decrypt_with_simulated_annealing,
        "MCMC (Metropolis-Hastings)": mcmc_decrypt
    }

    results = {method: {} for method in decryption_methods.keys()}
    times = {method: {} for method in decryption_methods.keys()}

    for order in range(1, 8):
        print(f"\nTesting Markov Chain of Order {order}:")
        model = markov_models[order]
        
        for method_name, method in decryption_methods.items():
            start_time = time.time()
            decrypted_text = method(cipher_text, model, order=order)
            elapsed_time = time.time() - start_time
            accuracy = evaluate_accuracy(decrypted_text, plaintext)
            
            results[method_name][order] = accuracy
            times[method_name][order] = elapsed_time

            print(f"\n{method_name} Results (Order {order}):")
            print(f"Decrypted text: {decrypted_text}")
            print(f"Accuracy: {accuracy * 100:.2f}%")
            print(f"Time: {elapsed_time:.2f} seconds")
            
    fig, ax1 = plt.subplots(figsize=(12, 6))

    ax2 = ax1.twinx()
    for method_name, accuracies in results.items():
        ax1.plot(accuracies.keys(), accuracies.values(), label=f'{method_name} Accuracy')
    for method_name, times in times.items():
        ax2.plot(times.keys(), times.values(), linestyle='--', label=f'{method_name} Time')

    ax1.set_xlabel('Markov Chain Order')
    ax1.set_ylabel('Accuracy', color='g')
    ax2.set_ylabel('Tim (seconds)', color='b')

    plt.title('Comparison of Decryption Methods')
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    plt.show()

test_decryption_methods()
```
